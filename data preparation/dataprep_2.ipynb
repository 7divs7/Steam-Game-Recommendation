DATA PREPARATION FOR COLLABORATIVE FILTERING - NeuCF

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
```

```python
df = pd.read_csv('../datasets/recommendations.csv')

df
```

```python
unique_user_ids = df['user_id'].unique()

first_2000_user_ids = unique_user_ids[:2000]

df = df[df['user_id'].isin(first_2000_user_ids)]

df
```

```python
df = df.drop(['helpful', 'funny', 'date', 'is_recommended', 'review_id'], axis=1)

df
```

```python
scaler = MinMaxScaler()
df['hours'] = scaler.fit_transform(df[['hours']])

df
```

```python
df.to_csv('../datasets/training_set_recommendations.csv', index=False)
```
